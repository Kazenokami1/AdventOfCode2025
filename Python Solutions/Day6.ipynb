```python
import math
with open("../Input Files/Day6.txt", encoding='utf-8') as file:
    my_data = file.read()
mathHomework = my_data.splitlines()

problems = []
for problem in mathHomework:
    problems.append(problem.split())

grandTotal = 0
for i in range(len(problems[0])):
    operation = problems[len(problems)-1][i]
    if operation == '*':
        problemSolution = 1
    else:
        problemSolution = 0
    for j in range(len(problems)-1):
        if operation == '*':
            problemSolution *= int(problems[j][i])
        else:
            problemSolution += int(problems[j][i])
    grandTotal += problemSolution
print("Part 1 Solution: ", grandTotal)

grandTotal = 0
columnIndicator = len(mathHomework[0])-1
while columnIndicator >= 0:
    columnCounter = 0
    numbers = []
    allSpaces = False
    operation = ''
    while not allSpaces:
        allSpaces = True
        number = ''
        for j in range(len(mathHomework)):
            if columnIndicator - columnCounter < 0:
                break
            charToCheck = mathHomework[j][columnIndicator-columnCounter]
            if charToCheck.isdigit():
                number += charToCheck
                allSpaces = False
            elif charToCheck != ' ':
                operation = charToCheck
        columnCounter += 1
        if not allSpaces:
            numbers.append(int(number))
        elif operation == '*':
            grandTotal += math.prod(numbers)
        else:
            grandTotal += sum(numbers)  
    columnIndicator -= columnCounter
    
print("Part 2 Solution: ", grandTotal)
```
