```python
with open("../Input Files/Day4.txt", encoding='utf-8') as file:
    my_data = file.read()
rolls = my_data.splitlines()

width = len(rolls[0])
height = len(rolls)

totalRolls = 0
rollsDictionary = dict()
for i in range(width):
    for j in range(height):
        rollsDictionary[tuple([i,j])]=rolls[i][j]
        if rolls[i][j] == '@':
            totalRolls += 1

accessibleRolls = 0
for i in range(width):
    for j in range(height):
        key = tuple([i,j])
        count = 0
        if key in rollsDictionary:
            if rollsDictionary[key]=='@':
                for k in range(-1,2):
                    for l in range(-1,2):
                        key2 = tuple([i+k,j+l])
                        if key2 in rollsDictionary:
                            if rollsDictionary[key2]=='@':
                                count += 1
                if count < 5: #5 instead of 4 because it will always find itself
                    accessibleRolls += 1

print("Part 1 Accessible Rolls: ", accessibleRolls)

previousTotalRolls = totalRolls+1
accessibleRolls = 0
while previousTotalRolls > totalRolls:
    previousTotalRolls = totalRolls
    for i in range(width):
        for j in range(height):
            key = tuple([i,j])
            count = 0
            if key in rollsDictionary:
                if rollsDictionary[key]=='@':
                    for k in range(-1,2):
                        for l in range(-1,2):
                            key2 = tuple([i+k,j+l])
                            if key2 in rollsDictionary:
                                if rollsDictionary[key2]=='@':
                                    count += 1
                    if count < 5: #5 instead of 4 because it will always find itself
                        accessibleRolls += 1
                        rollsDictionary[key] = '.'
                        totalRolls -= 1
            
print("Part 2 Accessible Rolls: ", accessibleRolls)
```
