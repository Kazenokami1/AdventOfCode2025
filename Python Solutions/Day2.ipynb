```python
with open("../Input Files/Day2.txt", encoding='utf-8') as file:
    my_data = file.read()
ids = my_data.split(",")
id = 0
invalidInputSum = 0
while id < len(ids):
    dashindex = ids[id].find("-")
    first = int(ids[id][0:dashindex])
    second = int(ids[id][dashindex+1:])
    for i in range(first, second+1):
        stringI = str(i)
        if len(stringI)%2!=0:
            continue
        half = int(len(stringI)/2)
        if stringI[0:half] == stringI[half:]:
            invalidInputSum += i
    id+=1
print("Sum of Invalid Inputs Part 1: ", invalidInputSum)

#Thoughts on Part 2
#From 1:half of the input length
#Check to see if the number is divisible by the number (1,2,3,etc)
#If it is, slice it into that many values (123123 would get broken into 1,2,3,1,2,3, then 12,31,23, then 123,123)
#Check if all of the sliced values are equal
#If they are, it is an invalid ID

id = 0
invalidInputSum = 0
while id < len(ids):
    dashindex = ids[id].find("-")
    first = int(ids[id][0:dashindex])
    second = int(ids[id][dashindex+1:])
    for testNumber in range(first, second+1):
        lenTest = len(str(testNumber))
        for i in range(1,int(lenTest/2)+1):
            if lenTest%i == 0:
                array = []
                for j in range(0, int(lenTest), i):
                    array.append(str(testNumber)[j:j+i])
                if len(set(array)) == 1:
                    invalidInputSum += testNumber
                    break
    id+=1
print("Sum of Invalid Inputs Part 2: ", invalidInputSum)
```
