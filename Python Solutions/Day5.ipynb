```python
with open("../Input Files/Day5.txt", encoding='utf-8') as file:
    my_data = file.read()
inventory = my_data.splitlines()

freshRanges = []
ingredientIds = []
for value in inventory:
    if value.__contains__('-'):
        dashIndex = value.find('-')
        freshRanges.append([int(value[0:dashIndex]), int(value[dashIndex+1:])])
    elif value != "":
        ingredientIds.append(int(value))

freshIngredients = 0
for id in ingredientIds:
    for range in freshRanges:
        if range[0] <= id <= range[1]:
            freshIngredients += 1
            break

print("Part 1 Fresh Ingredients: ", freshIngredients)

totalIngredients = 0
maxValue = float('-inf')
freshRanges.sort()
for range in freshRanges:
    bottomValue = max(range[0], maxValue+1) #Make sure we don't double count if the bottom value is the same as the max value
    topValue = max(range[1], maxValue) #This way could actually make the topValue be lower than the bottomValue but our next line ignores this scenario because topValue would equal maxValue
    if topValue != maxValue: #This avoids adding anything that is already fully encompassed
        totalIngredients += topValue - bottomValue + 1 #+1 because we need both sides added (2-1 actually needs to add 2 ingredients)
    maxValue = topValue
print("Part 2 Total Fresh Ingredient Numbers: ", totalIngredients)
```
