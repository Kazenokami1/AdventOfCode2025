```python
with open("../Input Files/Day3.txt", encoding='utf-8') as file:
    my_data = file.read()
batteryBanks = my_data.split()

joltageSum = 0
for bank in batteryBanks:
    firstIndex = 0
    secondIndex = 0
    for i in range(0,len(bank)):
        if bank[i] > bank[firstIndex] and i < len(bank)-1:
            firstIndex = i
            secondIndex = 0
        elif bank[i] > bank[secondIndex] or secondIndex == 0:
            secondIndex = i
    joltageSum += int(bank[firstIndex]+bank[secondIndex])
print("Part 1 Joltage Sum: ", joltageSum)

joltageSum = 0
for bank in batteryBanks:
    joltage = ""
    batteryLocation = -1
    for i in range(0,12):
        batteryRange = bank[batteryLocation+1:len(bank)+i-11]
        if len(batteryRange) == 1:
            joltage += bank[batteryLocation+1:]
            break
        subBatteryLocation = batteryRange.index(max(batteryRange))
        batteryLocation = batteryLocation+1+subBatteryLocation
        joltage += bank[batteryLocation]
    joltageSum += int(joltage)
print("Part 2 Joltage Sum: ", joltageSum)
```
