```python
with open("../Input Files/Day1.txt", encoding='utf-8') as file:
    my_data = file.read()
instructions = my_data.split()
i = 0
pointer = 50
zeroHits = 0
while i < len(instructions):
    if instructions[i][0] == 'R':
        pointer += int(instructions[i][1:])
    else:
        pointer -= int(instructions[i][1:])
    if pointer%100 == 0:
        zeroHits += 1
    i+=1
print("Solution to Part 1:", zeroHits)

i = 0
pointer = 50
zeroHits = 0
while i < len(instructions):
    previousPointer = pointer
    instruction = instructions[i]
    if instructions[i][0] == 'R':
        pointer += int(instructions[i][1:])
        while pointer >= 100:
            pointer -= 100
            zeroHits += 1
    else:
        pointer -= int(instructions[i][1:])
        while pointer <= 0:
            if pointer == 0:
                zeroHits += 1
                break
            else:
                pointer += 100
                zeroHits += 1
                if previousPointer == 0:
                    previousPointer = pointer
                    zeroHits -= 1
    i+=1
print("Solution to Part 2:", zeroHits)
```
