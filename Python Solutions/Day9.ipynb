```python
with open("../Input Files/Day9.txt", encoding='utf-8') as file:
    my_data = file.read()

floor = []
for row in my_data.split():
    tile = row.split(',')
    floor.append([int(tile[0]), int(tile[1])])
firstTile = my_data.split()[0].split(',')
floor.append([int(firstTile[0]), int(firstTile[1])])

maxRectangle = 0
for tile1 in floor[:-1]:
    for tile2 in floor[floor.index(tile1)+1:]:
        width = abs(int(tile1[1])-int(tile2[1]))+1
        height = abs(int(tile1[0])-int(tile2[0]))+1
        rectangleArea = width*height
        maxRectangle = max(maxRectangle, rectangleArea)

print("Part 1 Solution: ", maxRectangle)

#For Part 2: Check if each point in the perimeter of the rectangle is in the polygon

maxRectangle = 0
for tile1 in floor[:-1]:
    for tile2 in floor[floor.index(tile1)+1:]:
        width = abs(int(tile1[1])-int(tile2[1]))+1
        height = abs(int(tile1[0])-int(tile2[0]))+1
        rectangleArea = width*height
        if rectangleArea > maxRectangle:
            validRectangle = True
            top = min(tile1[1], tile2[1])
            bottom = max(tile1[1], tile2[1])
            left = min(tile1[0], tile2[0])
            right = max(tile1[0], tile2[0])
            for i in range(len(floor)-1):
                if left < floor[i][0] < right:
                    if min(floor[i][1], floor[i+1][1]) <= top < max(floor[i][1], floor[i+1][1]):
                        validRectangle = False
                        break
                    elif min(floor[i][1], floor[i+1][1]) < bottom <= max(floor[i][1], floor[i+1][1]):
                        validRectangle = False
                        break
                if top < floor[i][1] < bottom:
                    if min(floor[i][0], floor[i+1][0]) <= left < max(floor[i][0], floor[i+1][0]):
                        validRectangle = False
                        break
                    elif min(floor[i][0], floor[i+1][0]) < right <= max(floor[i][0], floor[i+1][0]):
                        validRectangle = False
                        break
            if validRectangle:
                maxRectangle = rectangleArea

print("Part 2 Solution: ", maxRectangle)
```
