```python
with open("../Input Files/Day8.txt", encoding='utf-8') as file:
    my_data = file.read()
junctions = my_data.split()

class junctionBox:
    def __init__(self, number, position):
        self.number = number
        self.position = position
        self.neighbors = []

junctionDict = dict()
junctionDistances = dict()
distances = []
for i in range(len(junctions)):
    position = junctions[i].split(',')
    junctionDict[i] = junctionBox(i, [int(position[0]), int(position[1]), int(position[2])])

for box1 in junctionDict.values():
    for box2 in junctionDict.values():
        if box1 != box2:
            distance = pow((pow((box1.position[0] - box2.position[0]),2) + pow((box1.position[1] - box2.position[1]),2) + pow((box1.position[2] - box2.position[2]),2)),0.5)
            if not junctionDistances.keys().__contains__(distance):
                distances.append(distance)
                junctionDistances[distance] = [box1, box2]

distances.sort()
circuits = []
for i in range(1000):
    circuits.append(junctionDistances[distances[0]])
    distances.remove(distances[0]) #Added this to shorten part 2, remove if hedging around niche scenario explained below

circuitsUnchanged = False
while not circuitsUnchanged:
    circuitsUnchanged = True
    for circuit1 in circuits:
        for circuit2 in circuits:
            if circuit1 != circuit2:
                for box in circuit1:
                    if circuit2.__contains__(box):
                        for box2 in circuit2:
                            if not circuit1.__contains__(box2):
                                circuit1.append(box2)
                        circuits.remove(circuit2)
                        circuitsUnchanged = False
                        break

circuits.sort(key=len, reverse=True)
longestCircuitsMultiplied = 1
for i in range(3):
    longestCircuitsMultiplied *= len(circuits[i])

print("Part 1 Solution: ", longestCircuitsMultiplied)

#In the most technical sense, if the Circuit was completed in the first 1000 connections, this would fail. I figured that was unlikely
#However if you want to hedge around that case, remove the line above that's mentioned for this, and add the following code:
#
#circuits.clear() -- Clear the circuits
#circuits.append(junctionDistances[distances[0]])
#distances.remove(distances[0])
#
#This increased my code execution time by about 4 seconds to have this.
lastJunctionBoxes = []
while len(distances) > 0:
    circuits.append(junctionDistances[distances[0]])
    circuitsUnchanged = False
    while not circuitsUnchanged:
        circuitsUnchanged = True
        for circuit2 in circuits[1:]:
            for box in circuit2:
                if circuits[0].__contains__(box):
                    for box2 in circuit2:
                        if not circuits[0].__contains__(box2):
                            circuits[0].append(box2)
                    circuits.remove(circuit2)
                    circuitsUnchanged = False
                    break
    if len(circuits[0]) == len(junctions):
        lastJunctionBoxes = junctionDistances[distances[0]]
        break
    distances.remove(distances[0])



print("Part 2 Junction Boxes", lastJunctionBoxes[0].position, lastJunctionBoxes[1].position)
print("Part 2 Solution: ", lastJunctionBoxes[0].position[0] * lastJunctionBoxes[1].position[0])
```
