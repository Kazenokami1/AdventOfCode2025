```python
with open("../Input Files/Day7.txt", encoding='utf-8') as file:
    my_data = file.read()
manifold = my_data.split()

class splitter:
    
    def __init__(self, location):
        self.location = location
        self.children = []
        self.pathsToEnd = 0

    def addChild(self, splitter):
        self.children.append(splitter)

    def calcPathsToEnd(self):
        if self.pathsToEnd == 0:
            if len(self.children) == 0:
                self.pathsToEnd = 2
            elif len(self.children) == 1:
                self.pathsToEnd = self.children[0].calcPathsToEnd() + 1
            else:
                for child in self.children:
                    self.pathsToEnd += child.calcPathsToEnd()
        return self.pathsToEnd

beamColumns = []
splitterDict = dict()

for i in range(len(manifold[0])):
    if manifold[0][i] == 'S':
        beamColumns.append(i)
        beamStart = splitter([0,i])
        break

beamSplits = 0
for j in range(1,len(manifold)):
    for i in range(len(manifold[1])):
        if manifold[j][i] == '^' and beamColumns.__contains__(i):
            splitterDict[tuple[j,i]] = splitter([j,i])
            while beamColumns.__contains__(i):
                beamColumns.remove(i)
            beamColumns.append(i-1)
            beamColumns.append(i+1)
            if beamSplits == 0:
                beamStart.addChild(splitterDict[tuple[j,i]])
            beamSplits += 1
print("Part 1 Solution: ", beamSplits)

for splitterNode in splitterDict.values():
    for i in range (-1,2,2):
        for j in range(splitterNode.location[0], len(manifold)):
            if splitterDict.keys().__contains__(tuple[j,splitterNode.location[1]+i]):
                splitterNode.addChild(splitterDict[tuple[j,splitterNode.location[1]+i]])
                break

print("Part 2 Solution: ", beamStart.children[0].calcPathsToEnd())
```
