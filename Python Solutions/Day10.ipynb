```python
with open("../Input Files/Day10Test.txt", encoding='utf-8') as file:
    my_data = file.read()
manual = my_data.splitlines()

lights = []
buttons = []
joltages = []
for row in manual:
    lights.append(row[1:row.index(']')])
    buttons.append(row[row.index(']')+2:row.index('{')-1])
    joltages.append(row[row.index('{')+1:row.index('}')])

valueMap = dict()
valueMap['.'] = '#'
valueMap['#'] = '.'

totalButtonsPressed = 0
for i in range(len(lights)):
    buttonPresses = float('inf')
    lightSequence = list(lights[i])
    panelLights = ['.']*len(lightSequence)
    panelButtons = buttons[i].replace('(', '').replace(')', '').split()
    binary = '0'
    while len(binary) <= len(panelButtons):
        binary = format(int(binary, 2) + 1, 'b')
        if binary.count('1') < buttonPresses:
            buttonsToPress = panelButtons[-len(binary):]
            for j in range(len(buttonsToPress)):
                if binary[j] == '1':
                    for light in buttonsToPress[j].split(','):
                        panelLights[int(light)] = valueMap[panelLights[int(light)]]
            if panelLights == lightSequence:
                buttonPresses = min(buttonPresses, binary.count('1'))
                panelLights = ['.']*len(lightSequence)
            else:
                panelLights = ['.']*len(lightSequence)
    totalButtonsPressed += buttonPresses

print("Part 1 Total Buttons Pressed:", totalButtonsPressed)

totalButtonsPressed = 0
for i in range(len(joltages)):
    joltageSequence = joltages[i].split(',')
    for j in range(len(joltageSequence)):
        joltageSequence[j] = int(joltageSequence[j])
    panelButtons = buttons[i].replace('(', '').replace(')', '').split()
    panelButtonMaxPushes = []
    for j in range(len(panelButtons)):
        panelButtonMaxPushes.append(float('inf'))
        for button in panelButtons[j].split(','):
            panelButtonMaxPushes[j] = min(panelButtonMaxPushes[j], joltageSequence[int(button)])
    currentJoltages = [0]*len(joltageSequence)

print(totalButtonsPressed)
```
