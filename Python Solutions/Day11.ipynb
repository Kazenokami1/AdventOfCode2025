```python
with open("../Input Files/Day11.txt", encoding='utf-8') as file:
    my_data = file.read()
devices = my_data.splitlines()

class device:

    def __init__(self, name):
        self.name = name
        self.children = []
        self.pathsToEnd = 0
        self.pathsToNode = dict()
        self.pathsToNode[name] = 1
    
    def addChild(self, name):
        self.children.append(name)

    def calcPathsToEnd(self):
        if self.pathsToEnd == 0:
            if len(self.children) == 0:
                self.pathsToEnd = 1
            else:
                for child in self.children:
                    self.pathsToEnd += child.calcPathsToEnd()
        return self.pathsToEnd
    
    def calcPathsToNode(self, nodeToFind):
        if not self.pathsToNode.keys().__contains__(nodeToFind):
            self.pathsToNode[nodeToFind] = 0
            for child in self.children:
                self.pathsToNode[nodeToFind] += child.calcPathsToNode(nodeToFind)
        return self.pathsToNode[nodeToFind]

deviceList = dict()
for listedDevice in devices:
    deviceList[listedDevice[:listedDevice.index(':')]] = device(listedDevice[:listedDevice.index(':')])

for listedDevice in devices:
    childDevices = listedDevice[listedDevice.index(':')+1:].split()
    for childDevice in childDevices:
        if not childDevice == 'out':
            deviceList[listedDevice[:listedDevice.index(':')]].addChild(deviceList[childDevice])

print("Part 1 Solution Total Paths Out: ", deviceList["you"].calcPathsToEnd())

with open("../Input Files/Day11.txt", encoding='utf-8') as file:
    my_data = file.read()
devices = my_data.splitlines()

deviceList2 = dict()
for listedDevice in devices:
    deviceList2[listedDevice[:listedDevice.index(':')]] = device(listedDevice[:listedDevice.index(':')])

for listedDevice in devices:
    childDevices = listedDevice[listedDevice.index(':')+1:].split()
    for childDevice in childDevices:
        if not childDevice == 'out':
            deviceList2[listedDevice[:listedDevice.index(':')]].addChild(deviceList2[childDevice])

deviceList2['svr'].calcPathsToEnd()
deviceList2['fft'].calcPathsToNode('dac')
deviceList2['dac'].calcPathsToNode('fft')
deviceList2['svr'].calcPathsToNode('dac')
deviceList2['svr'].calcPathsToNode('fft')
pathsFromSVRtoFFTtoDAC = deviceList2['svr'].pathsToNode['fft']*deviceList2['fft'].pathsToNode['dac']*deviceList2['dac'].pathsToEnd
pathsFromSVRtoDACtoFFT = deviceList2['svr'].pathsToNode['dac']*deviceList2['dac'].pathsToNode['fft']*deviceList2['fft'].pathsToEnd
print("Part 2 Solution Total Paths including dac and fft: ", max(pathsFromSVRtoFFTtoDAC, pathsFromSVRtoDACtoFFT))
```
